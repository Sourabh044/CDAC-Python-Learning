```python
import pandas as pd
import numpy as np
```

# Extract Adventure movies & sort them year-wise

```python
data = pd.read_csv('IMDB-Movie-Data.csv',index_col='Title')
```

```python
Query = data.sort_values(by=['Year'])
Query.loc[Query['Genre'].apply(lambda x:'Adventure' in x)]
```

# Extract Horror movies & sort them by rating

```python
Query = data.sort_values(by=['Rating'],ascending=False)
Query.loc[Query['Genre'].apply(lambda x:'Horror' in x)]
```

#  Calculate year-wise revenue from movies

```python
for x in sorted(data.Year.unique()):
    revY = data.loc[data.Year == x]
    print('Revenue generated in year',x,'by movies is',round(revY['Revenue (Millions)'].sum(),2))
```

# Extract Movie details with average runtime


```python
avg = data['Runtime (Minutes)'].mean()
print('Average Runtime is',avg,'Minutes')
avg_R_mov = data.loc[(data['Runtime (Minutes)']> avg-3) & (data['Runtime (Minutes)'] <avg + 1)] #range(110,114))]
avg_R_mov
```

# Lowest Votes movie in 2016


```python
Query = data.loc[data.Year==2016]
Query.loc[Query['Votes']== Query['Votes'].min()]
```

# Extracting title of movies of 'Vin Diesel' with year

```python
Query = data[['Actors','Year']]
Query.loc[Query['Actors'].apply(lambda x:'Vin Diesel' in x)]
```

# Calculate no of movies in genre-
# Action,Adventure,Sci-Fi,Comedy,Romance

```python
Query = data['Genre']
Query_list = ['Action','Adventure','Sci-Fi','Comedy','Romance']
for y in Query_list:
     a = Query.loc[Query.apply(lambda x:y in x)]
     print( 'Number of movies with genre',y,'is',a.count())
```
